```python
import time
import matplotlib.pyplot as plt

x = []
time_log = []
for i in range(50):
    tic = time.time()
    x.append(i)
    toc = time.time()
    time_log.append(toc-tic) 
    
plt.plot(time_log, '-*')
plt.show()
```
